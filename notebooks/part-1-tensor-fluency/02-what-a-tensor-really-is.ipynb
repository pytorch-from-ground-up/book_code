## The dimension ladder

A tensor is a box of numbers plus metadata. Rank is the number of axes; shape tells how many positions each axis contains.


```python
examples = {
    "scalar": torch.tensor(7),
    "vector": torch.tensor([1, 2, 3, 4]),
    "matrix": torch.tensor([[1, 2, 3], [4, 5, 6]]),
    "3-D tensor": torch.arange(24).reshape(2, 3, 4),
}
for name, value in examples.items():
    print(f"{name:10s} shape={tuple(value.shape)!s:10s} ndim={value.ndim}")
```

## Shape, dtype, device

These three properties explain most compatibility errors before an operation is attempted.


```python
batch = torch.randn(8, 3, 28, 28)
labels = torch.tensor([0, 1, 2, 1, 0, 2, 1, 0])
print("images:", batch.shape, batch.dtype, batch.device)
print("labels:", labels.shape, labels.dtype, labels.device)
print("batch, channels, height, width =", tuple(batch.shape))
```

## Layout and memory

Stride describes how storage is traversed. Number of elements multiplied by bytes per element gives the tensor's storage size.


```python
matrix = torch.arange(12, dtype=torch.float32).reshape(3, 4)
bytes_used = matrix.numel() * matrix.element_size()
print("shape:", matrix.shape, "stride:", matrix.stride())
print("elements:", matrix.numel(), "bytes:", bytes_used)
```
