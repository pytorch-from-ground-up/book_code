## Creation patterns

Choose constructors by intent: fixed values, ranges, evenly spaced values, random samples, or identity matrices.


```python
tensors = {
    "zeros": torch.zeros(2, 3),
    "ones": torch.ones(2, 3),
    "arange": torch.arange(0, 10, 2),
    "linspace": torch.linspace(0, 1, 5),
    "normal": torch.randn(2, 3),
    "identity": torch.eye(3),
}
for name, value in tensors.items():
    print(name, tuple(value.shape), value.dtype)
```

## Match an existing tensor

The *_like family preserves shape, dtype, and device, preventing subtle mismatches.


```python
reference = torch.randn(3, 4, dtype=torch.float64)
zeros = torch.zeros_like(reference)
noise = torch.randn_like(reference)
print(zeros.shape, zeros.dtype, zeros.device)
print("noise mean:", noise.mean().item())
```

## Copying and shared memory

torch.tensor copies input data; from_numpy can share memory with a NumPy array.


```python
import numpy as np

array = np.array([1.0, 2.0, 3.0], dtype=np.float32)
shared = torch.from_numpy(array)
copied = torch.tensor(array)
array[0] = 99
print("shared:", shared)
print("copied:", copied)
print("cast:", shared.to(torch.float64).dtype)
```
