## Build a tiny dataset

We learn a noisy line so the entire path from data to trained model stays visible.


```python
import torch.nn as nn

torch.manual_seed(0)
x = torch.linspace(0, 2, 40).unsqueeze(1)
y = 3 * x + 1 + 0.25 * torch.randn_like(x)
print(x.shape, y.shape)
```

## Model, loss, optimizer

The model predicts, the loss scores the prediction, and the optimizer owns the update rule.


```python
model = nn.Linear(1, 1)
loss_fn = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)
print(model)
```

## Train end to end

The order is stable: predict, measure, clear gradients, backpropagate, update.


```python
losses = []
for epoch in range(120):
    prediction = model(x)
    loss = loss_fn(prediction, y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    losses.append(loss.item())
print(f"loss: {losses[0]:.4f} -> {losses[-1]:.4f}")
print(f"learned y = {model.weight.item():.2f}x + {model.bias.item():.2f}")
```
