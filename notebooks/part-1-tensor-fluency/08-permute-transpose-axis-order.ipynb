## Transpose is a two-axis swap

For matrices, .T is concise. transpose(dim0, dim1) makes the swapped axes explicit.


```python
matrix = torch.arange(1, 9).reshape(2, 4)
swapped = matrix.transpose(0, 1)
print(matrix.shape, "->", swapped.shape)
print("contiguous:", swapped.is_contiguous(), "stride:", swapped.stride())
```

## Permute names the complete order

permute takes one position for every original axis. It is ideal when layouts have semantic names.


```python
tensor = torch.randn(2, 3, 4)
reordered = tensor.permute(2, 0, 1)
print("ABC", tensor.shape, "-> CAB", reordered.shape)
print("equivalent:", torch.equal(tensor.transpose(0, 2), tensor.permute(2, 1, 0)))
```

## Layouts in practice

Images often arrive as HWC but models expect CHW. Multi-head attention commonly moves the head axis before sequence.


```python
image_hwc = torch.randn(28, 28, 3)
image_chw = image_hwc.permute(2, 0, 1)
attention = torch.randn(2, 10, 4, 16)  # batch, sequence, heads, width
attention = attention.permute(0, 2, 1, 3)
print("image:", image_hwc.shape, "->", image_chw.shape)
print("attention:", attention.shape)
```
