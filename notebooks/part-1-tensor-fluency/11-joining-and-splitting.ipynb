## Grow an axis or create one

cat grows an existing axis; stack inserts a new axis and requires identical input shapes.


```python
a = torch.arange(1, 7).reshape(2, 3)
b = torch.arange(7, 13).reshape(2, 3)
print("cat rows:", torch.cat([a, b], dim=0).shape)
print("cat cols:", torch.cat([a, b], dim=1).shape)
print("stack:", torch.stack([a, b], dim=0).shape)
```

## Create a batch

A list of individual samples becomes a batch by stacking along a new leading axis.


```python
images = [torch.randn(3, 8, 8) for _ in range(5)]
batch = torch.stack(images, dim=0)
print("one image:", images[0].shape, "batch:", batch.shape)
```

## Chunk and split

chunk asks for a number of pieces; split asks for piece sizes. Both can be joined again with cat.


```python
tensor = torch.arange(20).reshape(5, 4)
chunks = torch.chunk(tensor, chunks=3, dim=0)
exact = torch.split(tensor, [2, 1, 2], dim=0)
print("chunk shapes:", [tuple(x.shape) for x in chunks])
print("split shapes:", [tuple(x.shape) for x in exact])
print("round trip:", torch.equal(torch.cat(exact, dim=0), tensor))
```
