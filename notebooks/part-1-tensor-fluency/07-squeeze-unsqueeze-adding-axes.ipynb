## Rows and columns from one vector

unsqueeze inserts an axis without copying data. The axis position determines later broadcasting behavior.


```python
vector = torch.tensor([1, 2, 3, 4])
row = vector.unsqueeze(0)
column = vector.unsqueeze(1)
print("vector", vector.shape, "row", row.shape, "column", column.shape)
print("same storage:", vector.data_ptr() == row.data_ptr())
```

## Squeeze with intent

Bare squeeze removes every size-one axis, which can accidentally remove a batch dimension. Prefer specifying the axis you understand.


```python
tensor = torch.randn(1, 3, 1, 4)
print("all:", tensor.squeeze().shape)
print("dim 0:", tensor.squeeze(0).shape)
print("dim 2:", tensor.squeeze(2).shape)
```

## Real shape preparation

Add a batch axis to one image, or reshape channel means to broadcast across height and width.


```python
sample = torch.randn(3, 28, 28)
batch = sample.unsqueeze(0)
means = torch.tensor([0.485, 0.456, 0.406])[:, None, None]
centered = sample - means
print("batch:", batch.shape, "means:", means.shape, "centered:", centered.shape)
```
