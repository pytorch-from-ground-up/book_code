## Basic indexing and slicing

PyTorch follows Python's zero-based indexing. A colon preserves an axis while an integer selects one position.


```python
matrix = torch.arange(20).reshape(4, 5)
print(matrix)
print("single:", matrix[1, 3].item())
print("column:", matrix[:, 2])
print("block:\n", matrix[1:3, 0:2])
print("reverse rows:\n", torch.flip(matrix, dims=(0,)))
```

## Masks and paired indices

A boolean mask filters by condition. Tensor indices can select specific coordinate pairs.


```python
print("greater than 12:", matrix[matrix > 12])
rows = torch.tensor([0, 1, 3])
cols = torch.tensor([4, 2, 0])
print("paired picks:", matrix[rows, cols])
```

## Views, clones, and gather

Basic slices usually share storage. Clone before editing when the source must remain unchanged. gather performs batched index selection.


```python
original = torch.arange(10)
view = original[:3]
safe = original[:3].clone()
view[0] = -1
safe[1] = -2
print("original:", original, "clone:", safe)

scores = torch.tensor([[0.1, 0.9, 0.3], [0.8, 0.2, 0.4]])
labels = torch.tensor([[1], [0]])
print("selected scores:", scores.gather(1, labels).squeeze(1))
```
