## A column plus a row

Dimensions align from the right. Each pair must match or one side must be 1.


```python
column = torch.tensor([[1], [2], [3]])
row = torch.tensor([[10, 20, 30, 40]])
result = column + row
print(column.shape, "+", row.shape, "=", result.shape)
print(result)
```

## Broadcast channel statistics

Shape channel statistics as (C, 1, 1) so they expand across spatial axes rather than across channels.


```python
image = torch.randn(3, 8, 8)
mean = torch.tensor([0.5, 0.4, 0.3])[:, None, None]
std = torch.tensor([0.2, 0.25, 0.3])[:, None, None]
normalized = (image - mean) / std
print(image.shape, mean.shape, normalized.shape)
```

## Catch a silent shape explosion

Compatible does not always mean intended. A vector (3,) and column (3,1) produce a 3 x 3 matrix.


```python
a = torch.randn(3)
b = torch.randn(3, 1)
surprising = a - b
intended = a - b.squeeze(1)
print("surprising:", surprising.shape, "intended:", intended.shape)
expanded = torch.ones(1, 4).expand(3, 4)
print("virtual expansion stride:", expanded.stride())
```
