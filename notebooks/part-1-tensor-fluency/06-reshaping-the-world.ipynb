## Same values, new arrangement

Reshaping changes the logical arrangement, not the values or element count.


```python
values = torch.arange(1, 13)
for shape in [(3, 4), (4, 3), (2, 6), (12,)]:
    arranged = values.reshape(shape)
    print(shape, "->", tuple(arranged.shape), "numel=", arranged.numel())
```

## Contiguity controls view

Transpose changes strides and often makes a tensor non-contiguous. reshape is the safe general choice; contiguous().view() makes copying explicit.


```python
base = torch.arange(6).reshape(2, 3)
transposed = base.T
print("base:", base.stride(), base.is_contiguous())
print("T:   ", transposed.stride(), transposed.is_contiguous())
print("reshape:", transposed.reshape(-1))
print("copy + view:", transposed.contiguous().view(-1))
```

## Infer and flatten dimensions

Use one -1 to infer a size. In models, flatten(1) keeps the batch dimension intact.


```python
tensor = torch.arange(24)
print(tensor.reshape(4, -1).shape)
batch = torch.randn(8, 3, 28, 28)
flat = batch.flatten(1)
print("before:", batch.shape, "after:", flat.shape)
assert flat.shape == (8, 3 * 28 * 28)
```
