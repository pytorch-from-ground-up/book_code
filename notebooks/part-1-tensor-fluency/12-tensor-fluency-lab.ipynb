## Checkerboard without loops

Two broadcasted coordinate vectors create every row-column pair at once.


```python
index = torch.arange(8)
board = ((index[:, None] + index[None, :]) % 2 == 0).long()
print(board)
assert board.shape == (8, 8)
```

## Standardize feature columns

Reduce over the sample axis and preserve it so each feature is centered and scaled independently.


```python
data = torch.randn(100, 5) * torch.tensor([1, 2, 3, 4, 5]) + 10
mean = data.mean(dim=0, keepdim=True)
std = data.std(dim=0, keepdim=True)
standardized = (data - mean) / std
print(standardized.mean(dim=0).round(decimals=4))
print(standardized.std(dim=0).round(decimals=4))
```

## Pairwise distances

Insert one axis on each copy of the point matrix, broadcast every pair, reduce feature differences, then take square roots.


```python
points = torch.randn(5, 3)
difference = points[:, None, :] - points[None, :, :]
distance = difference.square().sum(dim=2).sqrt()
print(distance.round(decimals=3))
print("symmetric:", torch.allclose(distance, distance.T))
print("zero diagonal:", torch.allclose(distance.diag(), torch.zeros(5)))
```
