        ## Learning objectives

        - Interpret a derivative as local slope
- Approximate slopes with finite differences
- Use gradients to move downhill


```python
import torch

torch.manual_seed(42)
print("PyTorch:", torch.__version__)
print("Default device: cpu")
```

## Slope from nearby points

A finite difference measures how much output changes after a very small input step.


```python
def f(x):
    return x ** 2 + 2 * x + 1

x = torch.tensor(3.0)
epsilon = 1e-4
numerical_slope = (f(x + epsilon) - f(x - epsilon)) / (2 * epsilon)
exact_slope = 2 * x + 2
print("numerical:", numerical_slope.item(), "exact:", exact_slope.item())
```

## A gradient points uphill

For many parameters, the gradient contains one local slope per parameter. Moving in the negative direction reduces the objective.


```python
point = torch.tensor([3.0, -2.0])
gradient = 2 * point  # gradient of x^2 + y^2
next_point = point - 0.1 * gradient
print("before:", point.square().sum().item())
print("after: ", next_point.square().sum().item())
```

## Manual gradient descent

Repeated small downhill steps approach the minimum. Learning rate controls the step size.


```python
weight = torch.tensor(8.0)
history = []
for _ in range(25):
    loss = (weight - 2) ** 2
    gradient = 2 * (weight - 2)
    weight = weight - 0.15 * gradient
    history.append(loss.item())
print("weight:", weight.item())
print("loss:", history[0], "->", history[-1])
```
