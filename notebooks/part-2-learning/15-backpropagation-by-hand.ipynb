```python
import torch

torch.manual_seed(42)
print("PyTorch:", torch.__version__)
print("Default device: cpu")
```

## Write the forward chain

Break a long formula into named intermediate values; this is the same graph PyTorch records.


```python
x_value = 2.0
a = x_value ** 2
b = 3 * a
c = b + 5
print({"x": x_value, "a=x^2": a, "b=3a": b, "c=b+5": c})
```

## Walk backward

Multiply local derivatives along the path: dc/db times db/da times da/dx.


```python
dc_db = 1.0
db_da = 3.0
da_dx = 2 * x_value
dc_dx = dc_db * db_da * da_dx
print("manual dc/dx:", dc_dx)
```

## Confirm and retain intermediate gradients

Non-leaf tensors do not keep .grad by default. retain_grad is useful for study and debugging.


```python
x = torch.tensor(2.0, requires_grad=True)
a = x ** 2
a.retain_grad()
c = 3 * a + 5
c.backward()
print("autograd dc/dx:", x.grad.item())
print("intermediate dc/da:", a.grad.item())
assert x.grad.item() == dc_dx
```
