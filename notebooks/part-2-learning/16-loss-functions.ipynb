## Regression losses

MSE penalizes large misses strongly; MAE measures absolute distance and is less sensitive to outliers.


```python
import torch.nn as nn

prediction = torch.tensor([2.5, 0.0, 4.0])
target = torch.tensor([3.0, -0.5, 2.0])
print("MSE:", nn.MSELoss()(prediction, target).item())
print("MAE:", nn.L1Loss()(prediction, target).item())
```

## Binary classification

BCEWithLogitsLoss combines a stable sigmoid calculation with binary cross-entropy. Targets must be floating-point values aligned with logits.


```python
logits = torch.tensor([-2.0, 0.5, 1.8], requires_grad=True)
binary_targets = torch.tensor([0.0, 1.0, 1.0])
binary_loss = nn.BCEWithLogitsLoss()(logits, binary_targets)
binary_loss.backward()
print("loss:", binary_loss.item(), "probabilities:", logits.sigmoid().detach())
```

## Multiclass classification

CrossEntropyLoss expects raw logits shaped (batch, classes) and integer class indices shaped (batch,). Do not apply softmax first.


```python
logits = torch.tensor([[2.0, 0.5, -1.0], [0.1, 0.2, 1.5]])
labels = torch.tensor([0, 2])
loss = nn.CrossEntropyLoss()(logits, labels)
print("loss:", loss.item())
print("predicted classes:", logits.argmax(dim=1))
```
