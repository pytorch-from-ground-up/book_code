## Autograd follows the computation

requires_grad tells PyTorch to remember operations so backward can apply the chain rule.


```python
x = torch.tensor(3.0, requires_grad=True)
y = x ** 2
z = 2 * y + 1
z.backward()
print("z:", z.item(), "dz/dx:", x.grad.item())
```

## Gradients accumulate

Each backward adds to the existing .grad buffer. Optimizers therefore clear gradients before the next batch.


```python
w = torch.tensor(2.0, requires_grad=True)
(w ** 2).backward()
first = w.grad.item()
(w ** 2).backward()
second = w.grad.item()
w.grad.zero_()
print("after one:", first, "after two:", second, "after clear:", w.grad.item())
```

## Stop tracking when appropriate

no_grad prevents graph construction for inference. detach creates a tensor view disconnected from the graph.


```python
source = torch.tensor([1.0, 2.0], requires_grad=True)
tracked = source * 3
detached = tracked.detach()
with torch.no_grad():
    inferred = source * 4
print(tracked.requires_grad, detached.requires_grad, inferred.requires_grad)
```
