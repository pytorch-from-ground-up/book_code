```python
import torch

torch.manual_seed(42)
print("PyTorch:", torch.__version__)
print("Default device: cpu")
```

## One explicit SGD step

An optimizer applies the same update you can write by hand: parameter minus learning rate times gradient.


```python
parameter = torch.tensor(5.0, requires_grad=True)
loss = (parameter - 1) ** 2
loss.backward()
with torch.no_grad():
    manual_next = parameter - 0.1 * parameter.grad
print("gradient:", parameter.grad.item(), "next:", manual_next.item())
```

## Compare optimizers fairly

Reset the same starting point for each optimizer so the comparison isolates the update rule.


```python
def optimize(kind, steps=40):
    value = torch.tensor([8.0], requires_grad=True)
    if kind == "sgd":
        optimizer = torch.optim.SGD([value], lr=0.1)
    elif kind == "momentum":
        optimizer = torch.optim.SGD([value], lr=0.1, momentum=0.9)
    else:
        optimizer = torch.optim.Adam([value], lr=0.2)
    history = []
    for _ in range(steps):
        loss = (value - 2).square().sum()
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        history.append(loss.item())
    return value.item(), history

for name in ["sgd", "momentum", "adam"]:
    value, history = optimize(name)
    print(f"{name:8s} value={value:6.3f} loss={history[-1]:.6f}")
```

## Learning-rate stress test

Too-small steps crawl; too-large steps oscillate or diverge.


```python
def manual_run(lr):
    value = 8.0
    losses = []
    for _ in range(12):
        losses.append((value - 2) ** 2)
        value -= lr * 2 * (value - 2)
    return value, losses

for lr in [0.01, 0.2, 1.1]:
    value, losses = manual_run(lr)
    print(f"lr={lr:<4} final={value:9.3f} last_loss={losses[-1]:.3f}")
```
