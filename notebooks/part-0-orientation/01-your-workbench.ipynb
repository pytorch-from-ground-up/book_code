```python
import torch

torch.manual_seed(42)
print("PyTorch:", torch.__version__)
print("Default device: cpu")
```

## Verify the workbench

A useful workbench reports its version and proves that tensor operations run.


```python
print("Version:", torch.__version__)
probe = torch.tensor([1, 2, 3]) + 1
print("Probe:", probe)
assert probe.tolist() == [2, 3, 4]
```

## Select a device once

Prefer CUDA on supported NVIDIA systems, then Apple's MPS backend, and always keep CPU as the dependable fallback.


```python
if torch.cuda.is_available():
    device = torch.device("cuda")
elif torch.backends.mps.is_available():
    device = torch.device("mps")
else:
    device = torch.device("cpu")
x = torch.arange(5, dtype=torch.float32).to(device)
print("Using:", device, "| tensor:", x)
```

## Control randomness

Resetting the seed reproduces the same pseudorandom sequence, which makes debugging and comparisons meaningful.


```python
torch.manual_seed(7)
first = torch.rand(4)
torch.manual_seed(7)
second = torch.rand(4)
print(first)
print("Reproduced:", torch.equal(first, second))
```
