## Learning curves

A train-validation pair reveals convergence, instability, and overfitting patterns that one scalar cannot show.


```python
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt

epochs = torch.arange(1, 21)
train_loss = torch.exp(-epochs / 7) + 0.05
val_loss = torch.exp(-epochs / 6) + 0.04 * torch.relu(epochs - 12)
fig, ax = plt.subplots(figsize=(6, 3))
ax.plot(epochs, train_loss, label="train")
ax.plot(epochs, val_loss, label="validation")
ax.set(xlabel="epoch", ylabel="loss", title="Learning curves")
ax.legend(); fig.tight_layout(); plt.close(fig)
print("created learning-curve figure")
```

## Predictions and confusion

For regression, compare predicted and actual values against the diagonal. For classification, annotate the confusion matrix.


```python
actual = torch.linspace(-2, 2, 40)
predicted = actual + 0.25 * torch.randn_like(actual)
confusion = torch.tensor([[18, 2], [4, 16]])
fig, axes = plt.subplots(1, 2, figsize=(8, 3))
axes[0].scatter(actual, predicted, s=15, alpha=0.7)
axes[0].plot([-2, 2], [-2, 2], "r--")
axes[0].set(xlabel="actual", ylabel="predicted")
image = axes[1].imshow(confusion, cmap="Blues")
for i in range(2):
    for j in range(2):
        axes[1].text(j, i, int(confusion[i, j]), ha="center", va="center")
axes[1].set(xlabel="predicted", ylabel="actual")
fig.colorbar(image, ax=axes[1]); fig.tight_layout(); plt.close(fig)
print("created prediction and confusion figures")
```

## Weights and activations

Hooks capture intermediate outputs without rewriting forward. Remove hooks after use to avoid repeated side effects.


```python
import torch.nn as nn

model = nn.Sequential(nn.Linear(4, 6), nn.ReLU(), nn.Linear(6, 2))
captured = {}
handle = model[1].register_forward_hook(lambda module, inputs, output: captured.update(relu=output.detach()))
_ = model(torch.randn(5, 4))
handle.remove()
print("first-layer weights:", model[0].weight.shape)
print("captured activations:", captured["relu"].shape)
```
