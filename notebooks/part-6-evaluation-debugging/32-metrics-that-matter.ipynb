```python
import torch

torch.manual_seed(42)
print("PyTorch:", torch.__version__)
print("Default device: cpu")
```

## Binary confusion counts

Every prediction falls into true positive, true negative, false positive, or false negative.


```python
target = torch.tensor([1, 0, 1, 1, 0, 0, 1, 0])
prediction = torch.tensor([1, 0, 0, 1, 1, 0, 1, 0])
tp = ((prediction == 1) & (target == 1)).sum()
tn = ((prediction == 0) & (target == 0)).sum()
fp = ((prediction == 1) & (target == 0)).sum()
fn = ((prediction == 0) & (target == 1)).sum()
confusion = torch.tensor([[tn, fp], [fn, tp]])
print(confusion)
```

## Precision, recall, F1

Precision asks how many predicted positives were right; recall asks how many actual positives were found. F1 balances both.


```python
precision = tp.float() / (tp + fp).clamp_min(1)
recall = tp.float() / (tp + fn).clamp_min(1)
f1 = 2 * precision * recall / (precision + recall).clamp_min(1e-12)
accuracy = (prediction == target).float().mean()
print({"accuracy": accuracy.item(), "precision": precision.item(), "recall": recall.item(), "f1": f1.item()})
```

## Regression metrics

MAE stays in target units, MSE emphasizes large errors, and R-squared compares against predicting the target mean.


```python
actual = torch.tensor([3.0, -0.5, 2.0, 7.0])
predicted = torch.tensor([2.5, 0.0, 2.0, 8.0])
error = predicted - actual
mae = error.abs().mean()
mse = error.square().mean()
r2 = 1 - error.square().sum() / (actual - actual.mean()).square().sum()
print("MAE:", mae.item(), "MSE:", mse.item(), "R2:", r2.item())
```
