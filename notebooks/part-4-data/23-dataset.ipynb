## TensorDataset first

When samples already live in tensors, TensorDataset provides length and indexing immediately.


```python
from torch.utils.data import Dataset, TensorDataset

features = torch.randn(12, 4)
labels = (features.sum(dim=1) > 0).long()
dataset = TensorDataset(features, labels)
print("length:", len(dataset), "one sample shapes:", dataset[0][0].shape, dataset[0][1].shape)
```

## A custom dataset

Implement __len__ and __getitem__. The index identifies one sample, not one batch.


```python
class StandardizedDataset(Dataset):
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.mean = x.mean(dim=0, keepdim=True)
        self.std = x.std(dim=0, keepdim=True).clamp_min(1e-6)

    def __len__(self):
        return len(self.x)

    def __getitem__(self, index):
        return (self.x[index] - self.mean.squeeze(0)) / self.std.squeeze(0), self.y[index]

custom = StandardizedDataset(features, labels)
x0, y0 = custom[0]
print(len(custom), x0.shape, y0.item())
```

## Inspect the sample contract

Models and DataLoaders depend on consistent sample types and shapes. Check several indices before training.


```python
for index in [0, len(custom) // 2, len(custom) - 1]:
    x_i, y_i = custom[index]
    print(index, tuple(x_i.shape), x_i.dtype, y_i.dtype)
```
