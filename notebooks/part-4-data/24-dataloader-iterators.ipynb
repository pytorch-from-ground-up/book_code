## From samples to batches

DataLoader repeatedly indexes the Dataset and collates matching fields along a new batch axis.


```python
from torch.utils.data import DataLoader, TensorDataset

x = torch.arange(40, dtype=torch.float32).reshape(10, 4)
y = torch.arange(10)
dataset = TensorDataset(x, y)
loader = DataLoader(dataset, batch_size=3, shuffle=False)
for batch_x, batch_y in loader:
    print(batch_x.shape, batch_y.tolist())
```

## Shuffle reproducibly

Pass a seeded Generator so experiments can reproduce the same first epoch order.


```python
def first_order(seed):
    generator = torch.Generator().manual_seed(seed)
    shuffled = DataLoader(dataset, batch_size=4, shuffle=True, generator=generator)
    return torch.cat([batch_y for _, batch_y in shuffled]).tolist()

print(first_order(7))
print("same:", first_order(7) == first_order(7))
```

## drop_last and iteration

The last batch may be smaller. drop_last is useful only when a model or normalization layer requires fixed batch size.


```python
keep = DataLoader(dataset, batch_size=4, drop_last=False)
drop = DataLoader(dataset, batch_size=4, drop_last=True)
print("batches kept:", len(keep), "batches dropped:", len(drop))
iterator = iter(keep)
first_x, first_y = next(iterator)
print(first_x.shape, first_y)
```
