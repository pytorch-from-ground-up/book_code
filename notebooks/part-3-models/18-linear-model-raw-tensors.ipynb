```python
import torch

torch.manual_seed(42)
print("PyTorch:", torch.__version__)
print("Default device: cpu")
```

## Data and raw parameters

A linear model needs only a weight and bias. Setting requires_grad makes both trainable leaves.


```python
torch.manual_seed(0)
x = torch.linspace(-2, 2, 60).unsqueeze(1)
y = 4 * x - 1 + 0.2 * torch.randn_like(x)
weight = torch.randn(1, 1, requires_grad=True)
bias = torch.zeros(1, requires_grad=True)
print(weight, bias)
```

## Manual forward and loss

Matrix multiplication plus a broadcasted bias is the complete forward pass.


```python
prediction = x @ weight + bias
loss = (prediction - y).square().mean()
print("initial loss:", loss.item())
```

## Manual training

After backward, update inside no_grad so the update itself is not added to the next computation graph, then clear gradients.


```python
for _ in range(160):
    prediction = x @ weight + bias
    loss = (prediction - y).square().mean()
    loss.backward()
    with torch.no_grad():
        weight -= 0.05 * weight.grad
        bias -= 0.05 * bias.grad
        weight.grad.zero_()
        bias.grad.zero_()
print(f"learned y = {weight.item():.2f}x {bias.item():+.2f}")
```
