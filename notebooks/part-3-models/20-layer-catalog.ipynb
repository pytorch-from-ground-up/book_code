## Core building blocks

Linear changes the final feature width; activations add nonlinearity; normalization and dropout affect training dynamics.


```python
import torch.nn as nn

model = nn.Sequential(
    nn.Linear(16, 32),
    nn.BatchNorm1d(32),
    nn.ReLU(),
    nn.Dropout(0.25),
    nn.Linear(32, 4),
)
x = torch.randn(8, 16)
print(model)
print("output:", model(x).shape)
```

## Shape tracing

Walk layer by layer when learning or debugging. Most layers preserve the batch axis.


```python
value = x
for layer in model:
    before = tuple(value.shape)
    value = layer(value)
    print(f"{layer.__class__.__name__:12s} {before} -> {tuple(value.shape)}")
```

## Embedding and flatten

Embedding turns integer IDs into dense vectors. Flatten combines selected axes before a linear projection.


```python
embedding = nn.Embedding(num_embeddings=20, embedding_dim=6)
token_ids = torch.tensor([[1, 4, 2], [3, 3, 5]])
embedded = embedding(token_ids)
flat = nn.Flatten(start_dim=1)(embedded)
projection = nn.Linear(flat.shape[1], 5)
print(token_ids.shape, "->", embedded.shape, "->", flat.shape, "->", projection(flat).shape)
```
