## Initialization by layer type

Kaiming initialization suits ReLU networks; biases are commonly set to zero. Apply functions visit every submodule.


```python
import torch.nn as nn

model = nn.Sequential(nn.Linear(10, 32), nn.ReLU(), nn.Linear(32, 4))

def initialize(module):
    if isinstance(module, nn.Linear):
        nn.init.kaiming_normal_(module.weight, nonlinearity="relu")
        nn.init.zeros_(module.bias)

model.apply(initialize)
for name, parameter in model.named_parameters():
    print(f"{name:10s} shape={tuple(parameter.shape)} mean={parameter.mean().item():+.3f}")
```

## Count and freeze

requires_grad controls whether autograd and the optimizer update a parameter.


```python
total = sum(p.numel() for p in model.parameters())
trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)
print("total:", total, "trainable:", trainable)
for parameter in model[0].parameters():
    parameter.requires_grad_(False)
print("after freezing first layer:", sum(p.numel() for p in model.parameters() if p.requires_grad))
```

## State dictionaries

A state_dict is a mapping from stable names to tensors. Loading it copies values into an identical architecture.


```python
state = {name: value.detach().clone() for name, value in model.state_dict().items()}
restored = nn.Sequential(nn.Linear(10, 32), nn.ReLU(), nn.Linear(32, 4))
restored.load_state_dict(state)
model.eval(); restored.eval()
sample = torch.randn(2, 10)
print("same output:", torch.allclose(model(sample), restored(sample)))
```
