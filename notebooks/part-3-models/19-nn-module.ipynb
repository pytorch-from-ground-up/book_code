        ## Learning objectives

        - Define a model as an nn.Module
- Register submodules and parameters
- Use call syntax instead of invoking forward directly


```python
import torch

torch.manual_seed(42)
print("PyTorch:", torch.__version__)
print("Default device: cpu")
```

## A minimal module

Assign layers in __init__, describe tensor flow in forward, and let the base class register everything.


```python
import torch.nn as nn

class LinearRegressor(nn.Module):
    def __init__(self, in_features=1):
        super().__init__()
        self.linear = nn.Linear(in_features, 1)

    def forward(self, x):
        return self.linear(x)

model = LinearRegressor()
print(model)
print(model(torch.tensor([[2.0]])))
```

## Registered state

parameters powers optimization; state_dict gives stable names for saving and loading.


```python
for name, parameter in model.named_parameters():
    print(name, tuple(parameter.shape), parameter.requires_grad)
print("state keys:", list(model.state_dict()))
```

## Train through the module

The training loop is unchanged because modules are callable tensor functions.


```python
x = torch.linspace(-1, 1, 40).unsqueeze(1)
y = 2.5 * x + 0.75
optimizer = torch.optim.SGD(model.parameters(), lr=0.15)
loss_fn = nn.MSELoss()
for _ in range(100):
    loss = loss_fn(model(x), y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
print(model.linear.weight.item(), model.linear.bias.item())
```
