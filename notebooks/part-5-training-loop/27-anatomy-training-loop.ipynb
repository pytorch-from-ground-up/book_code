## Prepare a complete problem

A TensorDataset and DataLoader let the loop focus on learning rather than sample assembly.


```python
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset

torch.manual_seed(0)
x = torch.randn(320, 10)
true_weight = torch.randn(10, 1)
y = x @ true_weight + 0.3 * torch.randn(320, 1)
loader = DataLoader(TensorDataset(x, y), batch_size=32, shuffle=True)
model = nn.Linear(10, 1)
loss_fn = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.03)
```

## The seven-step loop

For each batch: receive data, predict, measure, clear gradients, backpropagate, update, and account for the result.


```python
history = []
for epoch in range(15):
    epoch_loss = 0.0
    for batch_x, batch_y in loader:
        prediction = model(batch_x)
        loss = loss_fn(prediction, batch_y)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item() * len(batch_x)
    history.append(epoch_loss / len(loader.dataset))
print("loss:", round(history[0], 4), "->", round(history[-1], 4))
```

## Verify the learned relationship

Training loss is useful, but parameter recovery and predictions provide concrete checks for this synthetic problem.


```python
cosine = torch.nn.functional.cosine_similarity(model.weight.flatten(), true_weight.flatten(), dim=0)
with torch.no_grad():
    sample_loss = loss_fn(model(x[:16]), y[:16])
print("weight-direction similarity:", cosine.item())
print("sample loss:", sample_loss.item())
```
